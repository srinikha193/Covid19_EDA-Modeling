```python
#understadning cases of india ,china and japan comparison plot
df_india=df[df['location']=='India']
df_china=df[df['location']=='China']
df_japan=df[df['location']=='Japan']
```

```python
# prompt: #understadning cases of india ,china and japan comparison plot

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.lineplot(x='date',y='total_cases',data=df_india,label='India')
sns.lineplot(x='date',y='total_cases',data=df_china,label='China')
sns.lineplot(x='date',y='total_cases',data=df_japan,label='Japan')
plt.show()
```

```python
# prompt: also do a bar plot with good colors and neat one for the above comparison

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(x='date',y='total_cases',data=df_india,label='India')
sns.barplot(x='date',y='total_cases',data=df_china,label='China')
sns.barplot(x='date',y='total_cases',data=df_japan,label='Japan')
plt.legend()
plt.show()
```

```python
# prompt: #finding 3 countries with least number of cases,sum the total cases across all dates for a country so far and list them all out to date i need country name

df_grouped = df.groupby('location')['total_cases'].sum().reset_index()
df_sorted = df_grouped.sort_values('total_cases')
least_cases_countries = df_sorted.tail(20)
print(least_cases_countries)
```

```python
#getting the most recent dates data from the df
most_recent_date = df['date'].max()
df_most_recent = df[df['date'] == most_recent_date]
df_most_recent
```

```python
df_grouped = df.groupby('location')['total_cases'].sum().reset_index()
df_sorted = df_grouped.sort_values('total_cases', ascending=False)
top_15_countries = df_sorted.head(15)

plt.figure(figsize=(15, 10))
plt.xticks(rotation=90)
sns.barplot(x='location', y='total_cases', data=top_15_countries)
plt.title('Top 15 Countries with Most Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

```python

valid_countries = ['United States', 'India', 'France', 'Germany', 'Brazil', 'United Kingdom', 'Russia', 'Turkey', 'Italy', 'Spain', 'Argentina', 'Colombia', 'Mexico', 'Poland', 'Iran']
top_15_countries = top_15_countries[top_15_countries['location'].isin(valid_countries)]

plt.figure(figsize=(15, 10))
plt.xticks(rotation=90)
sns.barplot(x='location', y='total_cases', data=top_15_countries)
plt.title('Top 15 Countries with Most Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

```python

# Scatter plot for new cases vs. new deaths (for a specific country, e.g., India)
plt.figure(figsize=(15, 10))
sns.scatterplot(x='new_cases', y='new_deaths', data=df[df['location'] == 'India'])
plt.title('New Cases vs. New Deaths in India')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.show()
```

```python
# Distribution of total cases (histogram)
plt.figure(figsize=(15, 10))
sns.histplot(df['total_cases'], bins=50, kde=True)
plt.title('Distribution of Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Frequency')
plt.show()
```
